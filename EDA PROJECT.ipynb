```python
# NOW TRAINING A LINEAR REGRESSION MODEL
```

```python
# Considering X and Y Arrays 
```

```python
X = data.drop(['NOx(GT)','T','Time'], axis=1)

y= data['NOx(GT)']
```

```python
# USING Train Test Split
```

```python

pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Creating and Training the Model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
# NOW Model Evaluation
```

```python
print(lm.intercept_)
```

```python
coeff_data = pd.DataFrame(lm.coef_, index=X.columns, columns=['Coefficient'])
coeff_data
```

```python
# PREDICTION MODEL
```

```python
prediction = lm.predict(X_test)
```

```python
plt.scatter(y_test, prediction, c="purple", alpha=0.3)
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.title('Linear Regression Predicted vs Actual')
```

```python
linear_regression_score = lm.score(X_test, y_test)
linear_regression_score
```

```python
# Residualt Histogram
```

```python
sns.distplot((y_test-prediction), bins=70, color="purple")
```

```python
from sklearn import metrics
```

```python
print('MAE:',metrics.mean_absolute_error(y_test, prediction))
print('MSE:',metrics.mean_squared_error(y_test, prediction))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
